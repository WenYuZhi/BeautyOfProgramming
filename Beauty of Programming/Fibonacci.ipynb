```python
import time
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
ts = time.clock()
fibonacci(50)
print(time.clock() - ts)
```

```python
ts = time.clock()
n = 50
fib = [0 ,1]
while len(fib) <= n:
    fib.append(fib[-1] + fib[-2])
print(time.clock() - ts)
print(fib)
```
